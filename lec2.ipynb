```python
# 1. PYTHON OPERATORS

# Python has several types of operators that perform operations on variables and values.
# Let's explore each type with examples.

# 1.1 Arithmetic Operators
print("=== Arithmetic Operators ===")

a, b = 10, 3
print(f"a = {a}, b = {b}")
print(f"a + b = {a + b}")    # Addition
print(f"a - b = {a - b}")    # Subtraction
print(f"a * b = {a * b}")    # Multiplication
print(f"a / b = {a / b}")    # Division (returns float)
print(f"a // b = {a // b}")  # Floor Division (returns int, truncates decimal)
print(f"a % b = {a % b}")    # Modulus (remainder)
print(f"a ** b = {a ** b}")  # Exponentiation (power)
```

```python
# 1.2 Comparison Operators
print("\n=== Comparison Operators ===")

a, b = 10, 20
print(f"a = {a}, b = {b}")
print(f"a == b: {a == b}")  # Equal to
print(f"a != b: {a != b}")  # Not equal to
print(f"a > b: {a > b}")    # Greater than
print(f"a < b: {a < b}")    # Less than
print(f"a >= b: {a >= b}")  # Greater than or equal to
print(f"a <= b: {a <= b}")  # Less than or equal to

# Comparison operators return Boolean values (True or False)
# They are often used in conditional statements and loops


# take about if,input in python switch to slides 
```

```python
# 1.3 Logical Operators
print("\n=== Logical Operators ===")

x = True
y = False



print(F"x = {x}, y = {y}")
print(f"x and y: {x and y}")  # True if both are True
print(f"x or y: {x or y}")    # True if at least one is True
print(f"not x: {not x}")      # Inverts the boolean value

# Example with values instead of variables
print(f"5 > 3 and 10 > 5: {5 > 3 and 10 > 5}")  # Both True, so True
print(f"5 > 3 and 10 < 5: {5 > 3 and 10 < 5}")  # One False, so False
print(f"5 > 3 or 10 < 5: {5 > 3 or 10 < 5}")    # One True, so True

# Short-circuit evaluation: Python stops evaluating as soon as the result is determined
# In "and" operations, if first operand is False, second isn't evaluated
# In "or" operations, if first operand is True, second isn't evaluated
```

```python
# 1.4 Assignment Operators
print("\n=== Assignment Operators ===")

a = 10  # Simple assignment
print(f"Initial a = {a}")


a += 5   # Same as: a = a + 5
print(f"After a += 5: a = {a}")

a -= 3   # Same as: a = a - 3
print(f"After a -= 3: a = {a}")

a *= 2   # Same as: a = a * 2
print(f"After a *= 2: a = {a}")

a /= 4   # Same as: a = a / 4
print(f"After a /= 4: a = {a}")

a //= 2  # Same as: a = a // 2
print(f"After a //= 2: a = {a}")
a = a % 2
a %= 2   # Same as: a = a % 2
print(f"After a %= 2: a = {a}")

a = 2
a **= 3  # Same as: a = a ** 3
print(f"a = 2, after a **= 3: a = {a}")

userinput = int(input("Enter Number to check even or odd\n"))

if userinput % 2 == 0:
    print("Even")
else :
    print("Odd")
```

```python
# 1.5 Identity Operators
print("\n=== Identity Operators ===")

# "is" and "is not" check if two variables reference the same object in memory
# (not just if they have the same value)

a = [1, 2, 3]
b = [1, 2, 3]
c = a



print(f"a = {a}, b = {b}, c = {a}")
print(f"a is b: {a is b}")      # False (different objects, even though same values)
print(f"a is c: {a is c}")      # True (same object)
print(f"a is not b: {a is not b}")  # True
print(f"a == b: {a == b}")      # True (same values)

# Important distinction between "is" and "=="
# "==" checks if values are equal
# "is" checks if objects are exactly the same in memory

# Warning: For some small integers and strings, Python might reuse objects for optimization
x = 5
y = 5
print(f"x = {x}, y = {y}")
print(f"x is y: {x is y}")  # May be True due to Python's optimization

# For clarity, use "is" only with None, True, False, or when explicitly checking for same object
```

```python
# 1.6 Membership Operators
print("\n=== Membership Operators ===")

# "in" and "not in" check for membership in a sequence (string, list, tuple, set, dict)

fruits = ["apple", "banana", "cherry"]
print(f"fruits = {fruits}")
print(f"'apple' in fruits: {'apple' in fruits}")
print(f"'mango' in fruits: {'mango' in fruits}")
print(f"'mango' not in fruits: {'mango' not in fruits}")

text = "Hello World"
print(f"text = '{text}'")
print(f"'Hello' in text: {'Hello' in text}")
print(f"'Python' in text: {'Python' in text}")

# With dictionaries, "in" checks keys, not values
student = {"name": "John", "age": 20, "grade": "A"}
print(f"student = {student}")
print(f"'name' in student: {'name' in student}")  # Check for key
print(f"'John' in student: {'John' in student}")  # 'John' is a value, not a key

# To check for values in a dictionary
print(f"'John' in student.values(): {'John' in student.values()}")
```

```python
# 1.7 Bitwise Operators
print("\n=== Bitwise Operators ===")

# These operators work on bits and perform bit-by-bit operations
# They're used less frequently in everyday programming but important for certain tasks

a = 10  # Binary: 1010
b = 7   # Binary: 0111

print(f"a = {a} (binary: {bin(a)})")
print(f"b = {b} (binary: {bin(b)})")

# Bitwise AND
print(f"a & b = {a & b} (binary: {bin(a & b)})")  
# 1010 & 0111 = 0010 (decimal 2)

# Bitwise OR
print(f"a | b = {a | b} (binary: {bin(a | b)})")  
# 1010 | 0111 = 1111 (decimal 15)

# Bitwise XOR (exclusive OR)
print(f"a ^ b = {a ^ b} (binary: {bin(a ^ b)})")  
# 1010 ^ 0111 = 1101 (decimal 13)

# Bitwise NOT (flips all bits)
print(f"~a = {~a} (binary: {bin(~a)})")  
# ~1010 = -1011 (decimal -11) due to two's complement
# 10
# Left Shift (multiply by 2^n)
print(f"a << 1 = {a << 1} (binary: {bin(a << 1)})")  
# 1010 << 1 = 10100 (decimal 20)

# Right Shift (divide by 2^n)
print(f"a >> 1 = {a >> 1} (binary: {bin(a >> 1)})")  
# 1010 >> 1 = 0101 (decimal 5)

# Bitwise operators are useful for:
# - Flag operations
# - Low-level memory manipulation
# - Optimization in certain algorithms
```

```python
# 1.8 Operator Precedence
print("\n=== Operator Precedence ===")

# Operators have a precedence order (which ones get evaluated first)

# Example: arithmetic expression
result = (10 + 3) * 2 ** 2 - 8 / 4
      #= 10 + 3 * 4 - 8 / 4
      #= 10 + 12 - 8 / 4
      #= 10 + 12 - 2
print(f"10 + 3 * 2 ** 2 - 8 / 4 = {result}")

# Breakdown of calculation:
# 1. 2 ** 2 = 4 (exponentiation has highest precedence)
# 2. 3 * 4 = 12 (multiplication)
# 3. 8 / 4 = 2 (division)
# 4. 10 + 12 - 2 = 20 (addition and subtraction from left to right)

# Order of precedence (from highest to lowest):
# 1. Parentheses: ()
# 2. Exponentiation: **
# 3. Unary plus/minus: +x, -x
# 4. Multiplication, division, floor division, modulus: *, /, //, %
# 5. Addition, subtraction: +, -
# 6. Comparison operators: <, <=, >, >=, ==, !=
# 7. Logical NOT: not
# 8. Logical AND: and
# 9. Logical OR: or

# Using parentheses for clarity
result_with_parentheses = 10 + (3 * (2 ** 2)) - (8 / 4)
print(f"10 + (3 * (2 ** 2)) - (8 / 4) = {result_with_parentheses}")

# Changing precedence with parentheses
different_result = (10 + 3) * 2 ** (2 - 8) / 4
print(f"(10 + 3) * 2 ** (2 - 8) / 4 = {different_result}")


x = -800

if x :
    print(True)
else : 
    print(False)
```

```python
# 2. LOOPS IN PYTHON (FOR LOOP AND WHILE LOOP)

# Loops allow you to repeatedly execute a block of code

# 2.1 For Loop
print("=== For Loop ===")

# The for loop iterates over a sequence (like a list, tuple, string, etc.)

# Basic for loop with a list
print("Iterating through a list:")
fruits = ["apple", "banana", "cherry"]
for x,i in enumerate(fruits):
    print(f"I like {x} , {i}")
    print("ee")

# For loop with range function
print("\nUsing range(5):")
for i in range(5):  # range(5) generates numbers from 0 to 4
    print(f"Number: {i}")

# For loop with range - specifying start and end
print("\nUsing range(2, 6):")
for i in range(2, 6):  # range(2, 6) generates numbers from 2 to 5
    print(f"Number: {i}")

# For loop with range - specifying start, end, and step
print("\nUsing range(0, 10, 2):")
for i in range(0, 10, 2):  # range(0, 10, 2) generates 0, 2, 4, 6, 8
    print(f"Number: {i}")

# For loop with negative step
print("\nUsing range(10, 0, -2):")
for i in range(10, 0, -2):  # range(10, 0, -2) generates 10, 8, 6, 4, 2
    print(f"Number: {i}")

# For loop with a string (iterates through each character)
print("\nIterating through a string:")
for char in "Python":
    print(char)

# For loop with a dictionary
print("\nIterating through a dictionary:")
person = {"name": "John", "age": 30, "city": "New York"}

# By default, for loop iterates through keys of a dictionary
print("Keys:")
for key in person:
    print(key)

# Iterating through values
print("\nValues:")
for value in person.values():
    print(value)

# Iterating through key-value pairs
print("\nKey-value pairs:")
for key, value in person.items():
    print(f"{key}: {value}")
```

```python
# 2.2 Nested For Loops
print("\n=== Nested For Loops ===")

# Nested loops are loops inside other loops

# Example: Creating a simple multiplication table
print(":")
for i in range(1, 6):
    for j in range(1, 6):
       print(f'now i = {i} and j = {j}')

# Example: Finding all pairs from two lists
print("\nAll pairs from two lists:")
colors = ["red", "blue"]
sizes = ["S", "M", "L"]

for color in colors:
    for size in sizes:
        print(f"{color}-{size}", end=" ")
print()
```

```python
# 2.3 For Loop Control Statements
print("\n=== For Loop Control Statements ===")

# break: Exits the loop completely
print("Using break:")
for i in range(1, 10):
    if i == 5:
        print("Breaking at 5")
        break
    print(i, end=" ")
print()  # New line

# continue: Skips the current iteration and continues with the next
print("\nUsing continue:")
for i in range(1, 10):
    if i % 2 == 0:  # Skip even numbers
        continue
    print(i, end=" ")  # Only prints odd numbers

print()  # New line

# else clause with for loop
# The else clause executes after the loop completes normally (not if the loop is exited with break)
print("\nFor loop with else clause:")
for i in range(1, 5):
    print(i, end=" ")

print("- Loop completed successfully")

print("\nFor loop with break and else clause:")
for i in range(1, 10):
    print(i, end=" ")
    if i == 4:
        print("- Breaking at 4")
        break
else:
    print("This won't be printed because the loop was broken")
```

```python
# 2.4 While Loop
print("\n=== While Loop ===")

# While loop executes as long as the condition is True

# Simple while loop
print("Counting with while loop:")
count = 1
while count <= 5:
    print(count, end=" ")
    count += 1  # Increment counter (IMPORTANT: avoid infinite loops)
print()

# While loop with break
print("\nWhile loop with break:")
count = 1
while True:  # Infinite loop
    if count > 5:
        break  # Exit the loop when count exceeds 5
    print(count, end=" ")
    count += 1
print()

# While loop with continue
print("\nWhile loop with continue (skip even numbers):")
count = 0
while count < 10:
    count += 1
    if count % 2 == 0:
        continue  # Skip even numbers
    print(count, end=" ")
print()

# While loop with else
print("\nWhile loop with else:")
count = 1
while count <= 5:
    print(count, end=" ")
    count += 1
else:
    print("- Count is now", count)

# While loop with break and else
print("\nWhile loop with break and else:")
count = 1
while count <= 10:
    print(count, end=" ")
    if count == 5:
        break
    count += 1
else:
    print("This won't be printed because the loop was broken")
```

```python
# 3. PRE-TESTED AND POST-TESTED LOOPS

# In Python and programming in general, loops can be classified as either:
# 1. Pre-tested: The condition is checked BEFORE entering the loop body
# 2. Post-tested: The condition is checked AFTER executing the loop body

print("=== Pre-tested vs Post-tested Loops ===")

# 3.1 Pre-tested Loop (while loop in Python)
print("\n=== Pre-tested Loop (while) ===")
# The condition is checked before the loop body executes
# If the condition is False initially, the loop body may never execute

counter = 6  # Set initial value to fail the condition
print(f"Initial counter: {counter}")

print("Executing while loop (pre-tested)...")
while counter < 5:
    print(f"Loop body: counter = {counter}")
    counter += 1

print(f"After loop: counter = {counter}")
print("Since the condition was False initially, the loop body was never executed.")
```

```python
# 3.2 Post-tested Loop (do-while equivalent)
# Python doesn't have a direct do-while loop, but we can simulate one

print("\n=== Post-tested Loop (do-while emulation) ===")
# In a post-tested loop, the loop body executes at least once
# because the condition is checked AFTER the first execution

counter = 6  # Same initial value that would fail a pre-test
print(f"Initial counter: {counter}")

print("Executing do-while emulation (post-tested)...")
# Emulating do-while (post-tested) loop
while True:
    print(f"Loop body: counter = {counter}")
    counter += 1
    if not (counter < 5):  # Condition checked AFTER loop body executes
        break

print(f"After loop: counter = {counter}")
print("Even though the condition was already False, the loop body executed once.")

# 3.3 Another way to emulate a do-while loop
print("\n=== Alternative Post-tested Loop Emulation ===")

counter = 6
print(f"Initial counter: {counter}")

print("Executing alternative do-while emulation...")
# Another way to emulate a do-while loop
do_again = True
while do_again:
    print(f"Loop body: counter = {counter}")
    counter += 1
    do_again = counter < 5  # Determines whether to continue

print(f"After loop: counter = {counter}")

```

```python
# 3.4 For Loop as a Pre-tested Loop
print("\n=== For Loop (Pre-tested) ===")

# The for loop in Python is a pre-tested loop
# It checks if there are any items in the iterable before executing the loop body

empty_list = []
print("For loop with empty list:")
for item in empty_list:
    print("This will not be executed")
print("Loop finished without executing the body")

# With items, the loop executes
print("\nFor loop with non-empty list:")
for i in range(1, 4):
    print(f"Loop iteration: i = {i}")
```

```python
# 4. LOCAL AND GLOBAL VARIABLES

# In Python, variables can have different scopes - the region where they're accessible.
# The two main scopes are local and global.

print("\n=== LOCAL AND GLOBAL VARIABLES ===")

# 4.1 Global Variables
print("\n=== Global Variables ===")

# A global variable is defined in the main body of a Python script
# It's accessible throughout the entire script, including functions

# Defining a global variable
global_var = "I am a global variable"
print(f"Outside any function: {global_var}")

def access_global():
    # Accessing the global variable (read-only)
    print(f"Inside access_global() function: {global_var}")

access_global()
```

```python
# 4.2 Local Variables
print("\n=== Local Variables ===")

# A local variable is defined inside a function
# It's only accessible within that function

def local_variable_demo():
    # This is a local variable
    local_var = "I am a local variable"
    print(f"Inside function: {local_var}")

local_variable_demo()

# Uncommenting the next line would cause an error
# because local_var is not accessible outside the function
# print(local_var)  # This would raise NameError: name 'local_var' is not defined

# Another example with parameters (which are local variables)
def greet(name):  # 'name' is a local variable (parameter)
    message = f"Hello, {name}!"  # 'message' is a local variable
    print(message)

greet("Alice")
# print(name)     # NameError: name is not defined
# print(message)  # NameError: message is not defined
```

```python
# 4.3 Local Variables with Same Name as Global
print("\n=== Local Variables with Same Name as Global ===")

# If a local variable has the same name as a global variable,
# the local variable "shadows" the global one within the function

x = "global"
print(f"Global x: {x}")

def test_shadowing():
    x = "local"  # This creates a new local variable, not modifying the global x
    print(f"Inside function, local x: {x}")

test_shadowing()
print(f"Global x after function call: {x}")  # Global x is unchanged
```

```python
# 4.4 Modifying a Global Variable from a Function
print("\n=== Modifying Global Variables ===")

# By default, you can't modify a global variable from within a function
# You need to use the 'global' keyword

counter = 0  # Global variable
print(f"Initial counter: {counter}")

def increment_counter_wrong():
    # This creates a new local variable 'counter', not modifying the global one
    counter = counter + 1  # This would raise UnboundLocalError
    print(f"Inside function: {counter}")

# Uncommenting the next line would cause an error
# increment_counter_wrong()  # UnboundLocalError: local variable 'counter' referenced before assignment

def increment_counter_correct():
    global counter  # Declare that we want to use the global 'counter'
    counter = counter + 1
    print(f"Inside function: {counter}")

increment_counter_correct()
print(f"After function call: {counter}")  # Global counter is modified
```

```python
# 5. ARGUMENTS AND PARAMETERS IN PYTHON FUNCTIONS

print("\n=== ARGUMENTS AND PARAMETERS IN PYTHON FUNCTIONS ===")

# Parameters: Variables defined in a function declaration
# Arguments: Values passed to the function when it's called

# 5.1 Basic Parameters and Arguments
print("\n=== Basic Parameters and Arguments ===")

def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

greet("Alice")  # "Alice" is an argument
greet("Bob")    # "Bob" is an argument

# Multiple parameters
def describe_person(name, age, occupation):
    print(f"{name} is {age} years old and works as a {occupation}.")

describe_person("John", 30, "programmer")
```

```python
# 5.2 Default Parameters
print("\n=== Default Parameters ===")

# Default parameters are used when no value is provided
def greet_with_greeting(name, greeting="Hello"):  # 'greeting' has a default value
    print(f"{greeting}, {name}!")

greet_with_greeting("Alice")  # Uses default greeting
greet_with_greeting("Bob", "Hi")  # Overrides default greeting

# Default parameters must come after non-default parameters
# This is incorrect: def func(a=1, b): ...

# Example with multiple default parameters
def create_profile(name, age=25, city="Unknown", occupation="Unknown"):
    profile = {
        "name": name,
        "age": age,
        "city": city,
        "occupation": occupation
    }
    return profile

print(create_profile("Alice"))  # Using all defaults except name
print(create_profile("Bob", 30))  # Providing name and age
print(create_profile("Charlie", city="London"))  # Using keyword arguments
```

```python
# 5.3 Keyword Arguments
print("\n=== Keyword Arguments ===")

# When calling a function, you can specify arguments by parameter name

def display_info(name, age, city):
    print(f"Name: {name}, Age: {age}, City: {city}")

# Using positional arguments (order matters)
display_info("Alice", 30, "New York")

# Using keyword arguments (order doesn't matter)
display_info(city="London", name="Bob", age=25)

# Mixing positional and keyword arguments
# Positional arguments must come before keyword arguments
display_info("Charlie", city="Paris", age=35)

# This is incorrect: display_info(name="Dave", 40, "Berlin")
```

```python
# 5.4 Variable-Length Arguments (*args)
print("\n=== Variable-Length Arguments (*args) ===")

# *args allows a function to accept any number of positional arguments
# These arguments are collected into a tuple

def sum_all(*args):  # args is a tuple containing all arguments
    print(f"args: {args}, type: {type(args)}")
    total = 0
    for num in args:
        total += num
    return total

print(f"Sum of 1, 2: {sum_all(1, 2)}")
print(f"Sum of 1, 2, 3, 4, 5: {sum_all(1, 2, 3, 4, 5)}")
print(f"No arguments: {sum_all()}")  # Empty tuple

# *args with regular parameters
def team_info(team_name, captain, *members):
    print(f"Team: {team_name}")
    print(f"Captain: {captain}")
    print("Members:")
    for member in members:
        print(f"- {member}")

team_info("Python Devs", "Alice", "Bob", "Charlie", "Dave")
team_info("Data Scientists", "Eve")  # No additional members
```

```python
# NUMBER SYSTEMS AND CONVERSIONS IN PYTHON

print("\n=== NUMBER SYSTEMS AND CONVERSIONS ===")

# 1. Introduction to Number Systems
print("\n=== Introduction to Number Systems ===")

# Python supports different number systems:
# - Decimal (base 10): 0-9
# - Binary (base 2): 0-1
# - Octal (base 8): 0-7
# - Hexadecimal (base 16): 0-9, A-F

# Representation in Python:
decimal_num = 42        # Decimal (default)
binary_num = 0b101010   # Binary (prefix: 0b)
octal_num = 0o52        # Octal (prefix: 0o)
hex_num = 0x2A          # Hexadecimal (prefix: 0x)

print(f"Decimal: {decimal_num}")
print(f"Binary: {binary_num} (0b101010)")
print(f"Octal: {octal_num} (0o52)")
print(f"Hexadecimal: {hex_num} (0x2A)")

# All these represent the same value
print(f"All equal: {decimal_num == binary_num == octal_num == hex_num}")
```
